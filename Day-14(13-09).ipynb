```python
plt.figure()
plt.plot(x,y,'d-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Number Squares")
plt.show()
```

**Scatter plots**

```python
colors=['red','orange']*3
colors.extend(['green','blue'])
colors
```

```python
plt.figure()
x=[1,2,3,4,5,6,7,8]
y=x
plt.scatter(x,y,c=colors)
plt.show()
```

**Draw the plot x and y value(y=mx+c)**

```python

plt.figure()
x=[1,2,3,4,7,10]
y=[(2*i)+5 for i in x]
plt.plot(x,y,'-r')
```

```python
orange_width=[1,2,3,4,5]
orange_height=[4,5,6,7,8]
mango_height=[6,7,8]
mango_width=[8,9,10]
plt.figure()
plt.scatter(orange_width,orange_height,label="orange")
plt.scatter(mango_width,mango_height,label="mango")
plt.legend()
plt.show()
```
